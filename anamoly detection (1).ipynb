```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
import yfinance as yf
import seaborn as sns
```

```python
df = yf.Ticker("YESBANK.NS")
```

```python
hist = df.history(period="max")
```

```python
hist
```

```python
sns.boxplot(hist.index, hist.Volume)
```

```python
hist['month'] = hist.index.month
hist['year'] = hist.index.year
```

```python
sns.boxplot(hist.year, hist.Volume)
```

```python
hist_2018 = hist[hist.year == 2018]
hist_2019 = hist[hist.year == 2019]
hist_2020 = hist[hist.year == 2020]
```

```python
sns.boxplot(hist_2019.month, hist_2019.Volume)
```

```python
data0 = hist_2018.Volume
mean0 = np.mean(data0) 
# calculate standard deviation
sd0 = np.std(data0)
# create empty list to store outliers
outliers0 = []
# detect outlier
for i in data0: 
    z0 = (i-mean0)/sd0 # calculate z-score\
    if abs(z0) > 0.9:
        outliers0.append(1)
    else:
        outliers0.append(0) 
```

```python
hist_2018['outliers'] = outliers0
hist_2018
```

```python
data = hist_2019.Volume
mean = np.mean(data) 
# calculate standard deviation
sd = np.std(data)
# create empty list to store outliers
outliers = []
# detect outlier
for i in data: 
    z = (i-mean)/sd # calculate z-score\
    if abs(z) > 0.9:
        outliers.append(1)
    else:
        outliers.append(0) 
```

```python
hist_2019['outliers'] = outliers
hist_2019
```

```python
data1 = hist_2020.Volume
mean1 = np.mean(data) 
# calculate standard deviation
sd1 = np.std(data)
# create empty list to store outliers
outliers1 = []
# detect outlier
for i in data1: 
    z = (i-mean1)/sd1 # calculate z-score\
    if abs(z) > 0.9:
        outliers1.append(1)
    else:
        outliers1.append(0) 
        
hist_2020['outliers'] = outliers1
hist_2020
```

```python
frames = [hist_2018,hist_2019, hist_2020]
history = pd.concat(frames)
history
```

```python
plt.plot(history.index, history.outliers)
```

Logistic Regression

```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
```

```python
y = history.outliers
X_train, X_test, y_train, y_test = train_test_split(history, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
model = LogisticRegression(solver='liblinear', random_state=0).fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(model.score(X_test, y_test)))
```

Naive Bayes Model 

```python
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
accuracy = accuracy_score(y_test ,model.predict(X_test))*100
accuracy
```
